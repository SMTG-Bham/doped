### Analysis: $Cs_2SnI_6$, Intrinsic

```python
from doped.pycdt.utils.parse_calculations import get_vasprun
from doped.pycdt.core.chemical_potentials import UserChemPotAnalyzer
from doped import dope_stuff
```

Parsing chempots:

```python
# change this to point to your bulk:
bulk_vr = get_vasprun("PhaseDiagram/Cs2SnI6/Final_vstd/vasprun.xml") # provide a path to the computed bulk vasprun file
bulk_ce = bulk_vr.get_computed_entry()

cpa = UserChemPotAnalyzer(bulk_ce = bulk_ce, path_base="./") # path_base points to the folder with the PhaseDiagram folder in it
cpa.read_phase_diagram_and_chempots(include_mp_entries = False)
pd = cpa.phase_diagram
```

Save to pickle so don't have to rerun later:

```python
dope_stuff.save_to_pickle(pd, "PhaseDiagram/Cs2SnI6_pd.pickle")
```

```python
pd = dope_stuff.load_from_pickle("PhaseDiagram/Cs2SnI6_pd.pickle")
```

```python
print(pd.get_decomp_and_e_above_hull(bulk_ce, allow_negative=True)) # energy above hull is zero here
```

```python
print(f"{pd.get_phase_separation_energy(bulk_ce, ):.3f}") # phase separation energy
```

```python
print(f"{pd.get_form_energy_per_atom(bulk_ce):.3f}") # formation energy per atom
```

```python
print(f"{pd.get_form_energy(bulk_ce):.3f}") # formation energy per cell (one fu here)
```

```python
print(f"{pd.get_form_energy_per_atom([ent for ent in pd.stable_entries if ent.composition.reduced_formula == 'CsSnI3'][0]):.3f}") # formation energy per atom
```

```python
# Plot!
from pymatgen.analysis.phase_diagram import PDPlotter
plotter = PDPlotter(pd, show_unstable=False)  # you can also try show_unstable=True
plotter.show()
# plotter.write_image("{}.png".format('-'.join(system)), "png")  # save figure


print('Stable Entries (formula)\n--------')
for e in pd.stable_entries:
    print(e.composition.reduced_formula)

print('\nUnstable Entries (formula, e_above_hull (eV/atom), decomposes_to)\n--------')
for e in pd.unstable_entries:
    decomp, e_above_hull = pd.get_decomp_and_e_above_hull(e)
    pretty_decomp = ["{}:{}".format(k.composition.reduced_formula, round(v, 2)) for k, v in decomp.items()]
    print(e.composition.reduced_formula, "%.3f" % e_above_hull, pretty_decomp)
```

```python
for stable_entry in pd.stable_entries:  # CPLAP format
    print(len(stable_entry.composition.reduced_composition.as_dict()))
    for el_symbol, el_num in stable_entry.composition.reduced_composition.as_dict().items():
        print(int(el_num), el_symbol, end=" ")
    reduction_factor = el_num/stable_entry.composition.as_dict()[el_symbol]
    #print(stable_ent.energy*reduction_factor)
    print(pd.get_form_energy(stable_entry)*reduction_factor)
```
